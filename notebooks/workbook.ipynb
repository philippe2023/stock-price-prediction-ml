```python
def download_fucking_data(tickers):
    data = {}
    for ticker in tickers:
        print(f"Downloading data for {ticker}")
        stock_data = yf.download(ticker, start='2010-01-01', end='2020-12-31')
        if stock_data.empty:
            print(f"Sorry, data for {ticker} is not available")
        else:
            data[ticker] = stock_data
    return data
```

```python
stock_df = download_fucking_data(tickers)
stock_df
```

```python
for ticker in stock_df:
    stock_df[ticker].to_csv(f"../data/raw/{ticker}_data.csv")
```

```python
csv_folder_path = '../data/raw/'
```

```python
csv_files = glob.glob(os.path.join(csv_folder_path, '*.csv'))
```

```python
df_list = []

for file in csv_files:
    # Extract the ticker symbol from the filename (assuming format like 'AAPL_data.csv')
    ticker = os.path.basename(file).split('_')[0]  # Extract 'AAPL', 'MSFT', etc.
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # Add a 'Ticker' column to label the data with the stock symbol
    df['Ticker'] = ticker
    
    # Append the DataFrame to the list
    df_list.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

# Convert 'Date' to datetime if necessary
if 'Date' in combined_df.columns:
    combined_df['Date'] = pd.to_datetime(combined_df['Date'])

# Sort by 'Ticker' and 'Date' for easier readability
combined_df = combined_df.sort_values(by=['Ticker', 'Date'])

# Save the combined DataFrame to a new CSV file
# combined_df.to_csv('combined_stock_data.csv', index=False)

print("All files have been merged and saved to 'combined_stock_data.csv'.")
```
